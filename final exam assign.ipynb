## Step 1: Import a number of libraries that we expect to use

```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
import os
```

```python

print(os.getcwd())
```

```python
df = pd.read_csv('option1_dataset.csv')
```

## Data Preparation:

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop rows with more than 50% missing data
df.dropna(thresh=len(df.columns) * 0.5)
df
```

**Number of Rows:** The dataset contains 1,000 rows, where each row represents an instance or observation in the dataset.

**Number of Columns:** The dataset contains 22 columns (feature_1, feature_2, ..., feature_22), where each column represents a feature or attribute associated with the observations.

```python
# Drop columns with more than 70% missing data
df = df.loc[:, df.isnull().mean() < 0.7]
df
```

```python
## Impute missing values
numerical_columns = df.select_dtypes(include=["float64", "int64"]).columns
categorical_columns = df.select_dtypes(include=["object"]).columns
```

**Numerical Columns:** Missing values can be imputed with statistical measures like the median or mean.

**Categorical Columns:** Missing values can be filled with the most frequent value (mode) or a placeholder like "Unknown."

```python
# Impute numerical columns with the median
for col in numerical_columns:
    df[col] = df[col].fillna(df[col].median())

# Impute categorical columns with the mode
for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
## Encode categorical variables using one-hot encoding
dataset = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
## Standardize numerical variables
scaler = StandardScaler()
dataset[numerical_columns] = scaler.fit_transform(dataset[numerical_columns])
```

## # Step 2: Model Building

```python

## Separate features (X) and target (y)
X = dataset.iloc[:, :-1]  # Features
y = dataset.iloc[:, -1]   # Target
```

```python
## Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Ensure data is scaled
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Define the MLPClassifier
mlp = MLPClassifier(max_iter=1200, random_state=42)
```

*Supervised Learning Used for tasks where the goal is to predict class labels based on input features.

*Non-Linear Relationships Can handle datasets with non-linear decision boundaries due to its ability to learn complex patterns.

*Feature Interactions learns feature interactions through hidden layers.

## Step 3: Hyperparameter Tuning

```python

param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'sgd'],
    
}
```

### Document your hyperparameter selection process and justify your choices.


The following hyperparameters were selected for tuning to balance model performance, training time, and interpretability.

 **hidden_layer_sizes**
Specifies the number of neurons in each hidden layer. Each tuple represents the architecture of the network.
Values Tested:
(50,): A single hidden layer with 50 neurons.

(100,): A single hidden layer with 100 neurons.

(50, 50): Two hidden layers, each with 50 neurons.

**Justification:**
A smaller network (e.g., (50,)) is less complex, faster to train, and reduces overfitting on small datasets.
Larger or deeper networks (e.g., (100,) or (50, 50)) are tested to capture more complex relationships in the data.
Testing multiple configurations allows for identifying the best trade-off between model complexity and performance.

**activation**
Definition: Determines the activation function for neurons in the hidden layers.

Values Tested:

'relu' (Rectified Linear Unit): Efficient for deep networks and avoids vanishing gradients.

'tanh': Maps outputs between -1 and 1, which can help in scenarios with symmetric data distributions.

**Solver**
Definition: Determines the optimization algorithm for updating model weights during training.
Values Tested:

'adam': Adaptive Moment Estimation, which combines the benefits of RMSProp and momentum, is robust to different datasets and efficient for large datasets.

'sgd': Stochastic Gradient Descent, which is simpler but requires careful tuning of learning rates.
Justification:

'adam' is the default and widely used optimizer for its adaptive learning capabilities and faster convergence.
'sgd' is included to explore simpler optimizers and observe its behavior on the dataset.

**learning_rate**

Definition: Determines how the learning rate is adjusted during training.
Values Tested:

'constant': Keeps the learning rate fixed during training.

'adaptive': Reduces the learning rate if the training loss plateaus.

Justification:
'constant' allows testing a straightforward learning strategy.
'adaptive' is useful for datasets that require dynamic adjustment to converge effectively.

```python
grid_search = GridSearchCV(estimator=mlp, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)
grid_search.fit(X_train, y_train)
```

```python
# Step 3: Evaluation
## Best parameters from GridSearchCV
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

## Evaluate the best model on the test set
y_pred = best_model.predict(X_test)

# Print evaluation metrics
print("Best Parameters:", best_params)
print("Accuracy on Test Data:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
